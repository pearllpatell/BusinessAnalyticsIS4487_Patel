## Clean up the data
Did you notice anything unusual about the "house" amounts?

How about the handset price and income?

```python
df_clean = df[(df['house'] > 0) & (df['income'] > 0) & (df['handset_price'] < 1000)]
df_clean.describe()
```

# Add Derived Attribute

Add a column with the income in K (thousands)

```python
df_clean['income_k'] = (df_clean['income']/1000).round(2)
df_clean.head()
```

## Summarize the Data
 What is average house price at the two levels of leave?

 Conclusion?  Stayers have substantially more expensive homes.

```python
df_clean[['house','leave']].groupby('leave').mean().round(2)
```

## Data Visualization

➡️ Assignment Tasks
- Add at least two charts to show a summary of any major trends

```python
plt.hist(df['income'], bins=20, color='blue', edgecolor='black')
plt.title('Distribution of Income')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.show()
```

```python
average_price_per_satisfaction = df.groupby('reported_satisfaction')['handset_price'].mean()

average_price_per_satisfaction.plot(kind='bar', color='green')
plt.title('Average Handset Price by Reported Satisfaction')
plt.xlabel('Reported Satisfaction')
plt.ylabel('Average Handset Price')
plt.xticks(rotation=45)  # Rotating category labels for better readability
plt.show()
```
